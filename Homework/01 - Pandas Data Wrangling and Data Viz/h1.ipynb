**2.5—** Calculate the mean and the 95% confidence interval for the salary of left-handed and right-handed pitchers aggregating data across all years. Print the estimates. Additionally, present the estimates and the confidence intervals in a [point plot](https://seaborn.pydata.org/_images/pointplot_7_0.png). Calculate your confidence intervals using the standard error of the mean and ignoring missing values.

```python
import math
import seaborn as sns
import scipy.stats as stats
import warnings
warnings.filterwarnings('ignore') # ignore the warnings to make the notebook neater

# First filter out the data for left-handed and right-handed pitchers
left_salary = df.query("throws == 'L' and not salary.isna()")["salary"].values
right_salary = df.query("throws == 'R' and not salary.isna()")["salary"].values

# Calculate mean and CI for left-handed pitchers
mean_l, std_l = np.mean(left_salary), np.std(left_salary, ddof=1) # calculate the sample mean and sample standard deviation
se_l = std_l / math.sqrt(len(left_salary)) # calculate the standard error of the mean
print(f"The mean and the 95% confidence intercal for the salary of left-handed pitchers are: {mean_l:.2f} \
and [{mean_l - 1.96 * se_l:.2f}, {mean_l + 1.96 * se_l:.2f}]") # calculate the 95% confidence interval (1.96 is the z-score associated with the 95% confidence)

# Calculate mean and CI for right-handed pitchers
mean_r, std_r = np.mean(right_salary), np.std(right_salary, ddof=1)
se_r = std_r / math.sqrt(len(right_salary))
print(f"The mean and the 95% confidence intercal for the salary of right-handed pitchers are: {mean_r:.2f} \
and [{mean_r - 1.96 * se_r:.2f}, {mean_r + 1.96 * se_r:.2f}]")

# Plot the point plot and set 1.96*(standard error) for either side of the error bar
sns.pointplot(x="throws", y="salary", data=df[df["salary"].notnull()].loc[df['throws'].isin(['L','R'])], errorbar=('se', 1.96)).set(title="Mean of salary with 95% CI")
plt.show()
```

**2.7 —** Perform a t-test test comparing the salaries of left-handed and right-handed pitchers at the .05 level of significance. Consider data from all years. /Discuss:/ Who earns more?

```python
t_stat, p_value = stats.ttest_ind(df.query("throws == 'L' and not salary.isna()")["salary"].values,
                                  df.query("throws == 'R' and not salary.isna()")["salary"].values)
print(f"The t-statistic is: {t_stat}.")
print(f"The p-value is {p_value}.")
```

According to the p-value of the t-test result, we can reject the null hypothesis that there is no significant difference between the salaries of right-handed and left-handed pitchers. The t-statistic value is negative, indicating that the second group, i.e., the right-handed group, has larger mean value of salary. Therefore, we tend to believe right-handed pitchers earn more.

**3.4 —** /Discuss:/ Contrast the results obtained in Tasks 3.1 / 3.3 / 3.3 with the results obtained in Tasks 2.7 and 2.8. Are left-handed better than right-handed pitchers? Do they earn better salaries? Explain why the analysis might have reached different results.

Based on the results in tasks 3.3, we cannot observe a substantial distinction between the two groups due to significant overlap in their confidence intervals in both salary and BAOpp cases. This suggests the difference between left-handed pitchers and right-handed pitchers is not as statistically significant as what the p-values of t-test indicate in task 2.7/2.8.

To articulate this, we plot two new point plots of salaries and opponent's batting average(BAOpp) as in task 2.5. However, this time, we first calculate the means of salaries and the means of BAOpp for each group by year, and then, calculate the mean and the 95% condidence intervals of the averaged data across all the years.

```python
# Calculate the mean of salaries for the two groups by year
left_salary_per_year = df[df['throws']=='L'].groupby('yearID')['salary'].mean()
right_salary_per_year = df[df['throws']=='R'].groupby('yearID')['salary'].mean()

# Calculate the mean and standard error for both groups
mean_left_salary, mean_right_salary = left_salary_per_year.mean(), right_salary_per_year.mean()
se_left_salary, se_right_salary = left_salary_per_year.sem(), right_salary_per_year.sem()

# Using the critical value from the t-distribution (for large samples, the t-distribution approaches the normal distribution, so 1.96 is the appropriate z-score) to calculate the 95% confidence intervals
ci_left_salary = (mean_left_salary - 1.96*se_left_salary, mean_left_salary + 1.96*se_left_salary)
ci_right_salary = (mean_right_salary - 1.96*se_right_salary, mean_right_salary + 1.96*se_right_salary)

# Prepare data for plotting
data_salary = {
    'Handedness': ['Left-Handed', 'Right-Handed'],
    'Mean': [mean_left_salary, mean_right_salary],
    'Lower_CI': [ci_left_salary[0], ci_right_salary[0]],
    'Upper_CI': [ci_left_salary[1], ci_right_salary[1]]
}
df_plot_salary = pd.DataFrame(data_salary)

# Plot the point plot of means of salaries
plt.figure(figsize=(5, 3))
sns.pointplot(x='Handedness', y='Mean', data=df_plot_salary, join=False, capsize=0.1)
plt.errorbar(x=df_plot_salary['Handedness'], y=df_plot_salary['Mean'], 
             yerr=[df_plot_salary['Mean'] - df_plot_salary['Lower_CI'], df_plot_salary['Upper_CI'] - df_plot_salary['Mean']],
             fmt='o', capsize=5)
plt.title('Mean Salary per year with 95% Confidence Interval for Left-Handed vs Right-Handed Pitchers',y= 1.05)
plt.ylabel('Mean Salary')
plt.show()
```

```python
# Calculate the mean of BAOpp for the two groups by year
left_baopp_per_year = df[df['throws']=='L'].groupby('yearID')['BAOpp'].mean()
right_baopp_per_year = df[df['throws']=='R'].groupby('yearID')['BAOpp'].mean()

# Calculate the mean and standard error for both groups
mean_left_baopp, mean_right_baopp = left_baopp_per_year.mean(), right_baopp_per_year.mean()
se_left_baopp, se_right_baopp = left_baopp_per_year.sem(), right_baopp_per_year.sem()

# Using the critical value from the t-distribution (for large samples, the t-distribution approaches the normal distribution, so 1.96 is the appropriate z-score) to calculate the 95% confidence intervals
ci_left_baopp = (mean_left_baopp - 1.96*se_left_baopp, mean_left_baopp + 1.96*se_left_baopp)
ci_right_baopp = (mean_right_baopp - 1.96*se_right_baopp, mean_right_baopp + 1.96*se_right_baopp)

# Prepare data for plotting
data_baopp = {
    'Handedness': ['Left-Handed', 'Right-Handed'],
    'Mean': [mean_left_baopp, mean_right_baopp],
    'Lower_CI': [ci_left_baopp[0], ci_right_baopp[0]],
    'Upper_CI': [ci_left_baopp[1], ci_right_baopp[1]]
}
df_plot_baopp = pd.DataFrame(data_baopp)
# Plot using seaborn
plt.figure(figsize=(5, 3))
sns.pointplot(x='Handedness', y='Mean', data=df_plot_baopp, join=False, capsize=0.1)
plt.errorbar(x=df_plot_baopp['Handedness'], y=df_plot_baopp['Mean'], 
             yerr=[df_plot_baopp['Mean'] - df_plot_baopp['Lower_CI'], df_plot_baopp['Upper_CI'] - df_plot_baopp['Mean']],
             fmt='o', capsize=5)
plt.title('Mean BAOpp per year with 95% Confidence Interval for Left-Handed vs Right-Handed Pitchers',y= 1.05)
plt.ylabel('Mean BAOpp')
plt.show()
```

We can observe that the confidence intervals have a large overlap in both two plots, which is different from what we observe in task 2.5 where there is no overlapping.

```python
# T-test for salaries of left-handed and right-handed pitchers averaged by year
stats.ttest_ind(left_salary_per_year, right_salary_per_year)
```

```python
# T-test for BAOpp of left-handed and right-handed pitchers averaged by year
stats.ttest_ind(left_baopp_per_year, right_baopp_per_year)
```

We also perform the t-test for the same data and find that the the p-values are large (0.780 in the salary case and 0.199 in the BAOpp case), indicating the null hypothesis that two groups have no significant difference can't be rejected.

Why this analysis leads to different results? Note that t-test compares the means of two sample groups, we are actually comparing the means of salary and the means of BAOpp of the two groups in task 2.7/2.8 across all years. However, this approach raises two issues. 

First, as shown in task 3.3, Panel A, the proportion of left-handed pitchers to right-handed pitchers varies annually. That is to say, when calculating the average salary and BAOpp across all years, each year has different weights depending on the group. For instance, in the year with the smallest proportion of left-handed pitchers, such as 2001, right-handed pitcher data may far outweigh that of left-handed pitchers. Consequently, when calculating the average, 2001 would have a more significant impact on right-handed pitchers compared to left-handed pitchers.

Second, it can be clearly observed from task 3.2, Panel B, that the overall salary increases year by year, but the proportion of left-handers in the total population significantly decreases, as shown in task 3.2, Panel A. Therefore, right-handed pitchers tend to have more recent-year data than left-handed pitchers, hence increasing their average salary across all years. Such method of calculation is biased for left-handed pitchers, which does not accurately reflect the true difference between left-handed and right-handed players. Similarly, the trend of the proportion of left-handers does not exactly match the change in BAOpp, meaning that during the t-test, the t-value and p-value may be influenced by this imbalance.
